## 2. Ejercicio: Regresión lineal múltiple

En este ejercicio aprenderás a ajustar un modelo de regresión lineal múltiple para predecir una variable a partir de varias características. Usaremos un dataset de ejemplo y te dejaremos un espacio para que modifiques el código y experimentes.

```python
# Ejemplo: Regresión lineal múltiple con Boston Housing
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.linear_model import LinearRegression

# Cargar datos Boston Housing
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
columnas = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
df = pd.read_csv(url, delim_whitespace=True, names=columnas)
df.head()

# Usar varias variables predictoras
X = df[['RM', 'LSTAT', 'PTRATIO']]
y = df['MEDV']

modelo = LinearRegression()
modelo.fit(X, y)

print('Coeficientes:', modelo.coef_)
print('Intercepto:', modelo.intercept_)
```

Ahora es tu turno...

1. Cambia las variables que alimentan el modelo de regresión y observa cómo cambian los resultados.
2. Compara el desempeño respecto a otros modelos de regresión como la regresión de Ridge o Lasso. Aqui es importante que revise los parámetros de regularización y cómo afectan el ajuste del modelo.
3. Realiza una tabla comparativa de las métricas de desempeño de los modelos utilizados con el mejor resultado obtenido incluyendo una columna donde mencione las características utilizadas.
4. Elabora gráficas de salida que permitan comparar el desempeño de los diferentes modelos (muestra solo el mejor resultado).
5. Realiza una breve conclusión como análisis del impacto de las variables en el precio de la vivienda.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
columnas = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD",
            "TAX","PTRATIO","B","LSTAT","MEDV"]
df = pd.read_csv(url, delim_whitespace=True, names=columnas)

y = df["MEDV"]
X_temp = df.drop("MEDV", axis=1)
X_train_full, X_test_full, y_train, y_test = train_test_split(
    X_temp, y, test_size=0.2, random_state=42
)

combinaciones = {
    "Combo1": ["RM", "LSTAT", "PTRATIO"],
    "Combo2": ["LSTAT", "TAX", "PTRATIO", "DIS"],
    "Combo3": ["RM", "AGE", "RAD", "CHAS"],
    "Combo4": ["NOX", "CRIM", "INDUS", "DIS"],
    "Combo5": ["RM", "LSTAT", "PTRATIO", "NOX", "CRIM", "TAX", "DIS"],
    "Combo6": df.columns[:-1].tolist()
}

resultados_combos = []
for nombre, variables in combinaciones.items():
    X_train = X_train_full[variables]
    X_test = X_test_full[variables]
    modelo = LinearRegression().fit(X_train, y_train)
    y_pred = modelo.predict(X_test)
    resultados_combos.append({
        "Combinacion": nombre,
        "R2": r2_score(y_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
        "MAE": mean_absolute_error(y_test, y_pred),
        "Variables": variables
    })

df_combos = pd.DataFrame(resultados_combos).sort_values("R2", ascending=False)
print("Resultados por combinación de variables:\n", df_combos)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_full)
X_test_scaled = scaler.transform(X_test_full)

modelos = {
    "Linear": (LinearRegression(), {}, False),
    "Ridge": (Ridge(), {"alpha": [0.1, 1.0, 10.0]}, True),
    "Lasso": (Lasso(max_iter=2000), {"alpha": [0.01, 0.1, 1.0]}, True)
}

resultados_modelos = []
for nombre, (modelo, params, usar_scaler) in modelos.items():
    if usar_scaler:
        X_train_use, X_test_use = X_train_scaled, X_test_scaled
    else:
        X_train_use, X_test_use = X_train_full, X_test_full

    if params:
        grid = GridSearchCV(modelo, params, cv=5, scoring="r2")
        grid.fit(X_train_use, y_train)
        modelo = grid.best_estimator_
    else:
        modelo.fit(X_train_use, y_train)

    y_pred = modelo.predict(X_test_use)
    resultados_modelos.append({
        "Modelo": nombre,
        "R2": r2_score(y_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
        "MAE": mean_absolute_error(y_test, y_pred)
    })

df_modelos = pd.DataFrame(resultados_modelos).sort_values("R2", ascending=False)
print("\nComparación de modelos:\n", df_modelos)

mejor = df_modelos.iloc[0]["Modelo"]
if mejor == "Linear":
    X_train_use, X_test_use = X_train_full, X_test_full
elif mejor == "Ridge" or mejor == "Lasso":
    X_train_use, X_test_use = X_train_scaled, X_test_scaled

modelo_final = [m for m in modelos[mejor]][0].fit(X_train_use, y_train)
y_pred_final = modelo_final.predict(X_test_use)

plt.scatter(y_test, y_pred_final, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "r--")
plt.xlabel("Valores Reales")
plt.ylabel("Valores Predichos")
plt.title(f"Mejor modelo: {mejor}")
plt.show()
```

El modelo explica un 66% del precio de la vivienda con todas las variables, vi elegir muchas variables ayuda poco, aunque es precisamente por la disposicion del dataset. Por esto mismo linear, ridge y lasso nos dan resultados casi iguales, los datos están bastante limpios aunque nos sigue dando un error de entre 3 y 5 mil dolares, que es un factor determinante en caso que que utilizaramos un modelo como este para buscar vivienda, es bastante, sigue siendo un buen marco de referencia.

## 3. Ejercicio: Regresión para imputación de datos

En este ejercicio aprenderás a usar regresión para rellenar valores faltantes en un dataset. Usaremos un ejemplo y te dejaremos un espacio para que selecciones las variables y completes el proceso de imputación.

**Ejemplo:** Imputación de valores faltantes en consumo energético usando temperatura y humedad.

```python
# Ejemplo: Imputación de datos faltantes con regresión usando datos de energía
import pandas as pd
from sklearn.linear_model import LinearRegression

# Cargar datos de Appliances Energy (fragmento)
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
df = pd.read_csv(url)

# Seleccionar variables relevantes y simular valores faltantes
df = df[['T_out', 'RH_out', 'Appliances']].copy()
df.loc[10:20, 'Appliances'] = None

# Separar filas con y sin datos faltantes
df_train = df.dropna()
df_missing = df[df['Appliances'].isna()]

# Entrenar modelo para imputar consumo
X_train = df_train[['T_out', 'RH_out']]
y_train = df_train['Appliances']
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Imputar valores faltantes
X_missing = df_missing[['T_out', 'RH_out']]
df.loc[df['Appliances'].isna(), 'Appliances'] = modelo.predict(X_missing)
```

1. Revisa el dataset energy data (desde la línea donde se carga la base de datos) y genera un mapa de calor utilizando la correlación entre las características con respecto a la variable objetivo (por ejemplo, appliances).
2. Selecciona las variables que consideres más relevantes para predecir la variable objetivo.

```python
# Actividad: Regresión para imputación de datos
'''
1. Revisa el dataset energy data (desde la línea donde se carga la base de datos) y genera un mapa de calor utilizando la correlación entre las características con respecto a la variable objetivo (por ejemplo, appliances).
2. Selecciona las variables que consideres más relevantes para predecir la variable objetivo.
3. Cambia intervalo de datos faltantes.
4. Aplica diferentes modelos de regresión, y compara su desempeño.
5. Grafica los resultados obtenidos para cada modelo
'''
# Tu código aquí

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# 1. Cargar datos
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
df = pd.read_csv(url)

# 2. Seleccionar algunas variables
df_vars = df[['T_out','RH_out','Appliances','T2','RH_2','T6','RH_6']].copy()

plt.figure(figsize=(8,6))
sns.heatmap(df_vars.corr(), annot=True, cmap='coolwarm', center=0)
plt.title("Mapa de calor de correlaciones")
plt.show()

# 3. Simular datos faltantes en Appliances
df_vars.loc[100:150, 'Appliances'] = np.nan

df_train = df_vars.dropna()
df_missing = df_vars[df_vars['Appliances'].isna()]

X_train = df_train.drop('Appliances', axis=1)
y_train = df_train['Appliances']
X_missing = df_missing.drop('Appliances', axis=1)

# 4. Modelos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_missing_scaled = scaler.transform(X_missing)

modelos = {
    "Linear": (LinearRegression(), {}, False),
    "Ridge": (Ridge(), {"alpha":[0.1,1,10]}, True),
    "Lasso": (Lasso(max_iter=2000), {"alpha":[0.01,0.1,1]}, True)
}

resultados = []
imputaciones_dict = {}

for nombre,(modelo,params,usar_scaler) in modelos.items():
    X_use = X_train_scaled if usar_scaler else X_train

    if params:
        grid = GridSearchCV(modelo, params, cv=5, scoring="r2")
        grid.fit(X_use, y_train)
        modelo = grid.best_estimator_
    else:
        modelo.fit(X_use, y_train)

    y_pred = modelo.predict(X_use)
    resultados.append({
        "Modelo": nombre,
        "R2": r2_score(y_train, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_train, y_pred))
    })
    imputaciones = modelo.predict(X_missing_scaled if usar_scaler else X_missing)
    imputaciones_dict[nombre] = imputaciones

df_resultados = pd.DataFrame(resultados).sort_values("R2", ascending=False)
print(df_resultados)

# 5. Graficar comparación
plt.bar(df_resultados['Modelo'], df_resultados['R2'])
plt.ylabel("R² Score")
plt.title("Comparación de desempeño de modelos")
plt.show()

```

```python
# Actividad: Forecasting con regresión

'''
1. Cambia el valor de lag (por ejemplo, prueba lag de 7 o 15 días) y observa cómo cambia la predicción.
2. Prueba diferentes modelos de regresión (por ejemplo, DecisionTreeRegressor, Ridge, Lasso) y compara su desempeño.
3. Realiza una gráfica comparando las predicciones con los valores reales.
4. Calcula métricas de desempeño (MAE, RMSE, R2) para cada modelo y lag.
5. Escribe una breve conclusión sobre qué configuración y modelo predice mejor la temperatura.
'''
# Tu codigo aquí
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv'
df = pd.read_csv(url)

def crear_lag(df, lag=1):
    df_copy = df.copy()
    df_copy[f'Temp_lag{lag}'] = df_copy['Temp'].shift(lag)
    return df_copy.dropna()

def evaluar_modelo(modelo, X, y):
    modelo.fit(X, y)
    pred = modelo.predict(X)
    mae = mean_absolute_error(y, pred)
    rmse = np.sqrt(mean_squared_error(y, pred))
    r2 = r2_score(y, pred)
    return pred, mae, rmse, r2

lags = [1, 7, 15]
modelos = {
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.01),
    "DecisionTree": DecisionTreeRegressor(max_depth=5, random_state=42)
}

resultados = []
for lag in lags:
    df_lag = crear_lag(df, lag)
    X = df_lag[[f'Temp_lag{lag}']]
    y = df_lag['Temp']

    for nombre, modelo in modelos.items():
        pred, mae, rmse, r2 = evaluar_modelo(modelo, X, y)
        resultados.append({"Lag": lag, "Modelo": nombre, "MAE": mae, "RMSE": rmse, "R2": r2})

        if nombre == "LinearRegression" and lag == 1:
            plt.figure(figsize=(10, 5))
            plt.plot(df_lag['Date'], y, label="Real", alpha=0.7)
            plt.plot(df_lag['Date'], pred, label="Predicción", alpha=0.7)
            plt.title(f"Predicción de temperatura con {nombre} (lag={lag})")
            plt.xlabel("Fecha")
            plt.ylabel("Temperatura mínima")
            plt.legend()
            plt.show()

df_resultados = pd.DataFrame(resultados)
print("\nResultados comparativos:")
print(df_resultados)

mejor = df_resultados.sort_values("RMSE").iloc[0]
print("\nConclusión:")
print(f"El mejor modelo fue {mejor['Modelo']} con lag={mejor['Lag']}, "
      f"RMSE={mejor['RMSE']:.3f}, MAE={mejor['MAE']:.3f}, R2={mejor['R2']:.3f}")
```

En este caso con el lag de 1 todos los modelos tienden a comportarse mucho mejor. Podemos ver claramente como para el lag de 15 todos los modelos tienen basicamente la mitad del desempeño que sus contrapartes de 1 día de lag
